```python
# Testing accuracy of OCR operation for Shona_doc

pip install python-Levenshtein

from Levenshtein import distance

def calculate_ocr_accuracy(ocr_output, ground_truth):
    # Calculate character-level error rate (CER)
    char_distance = distance(ocr_output, ground_truth)
    char_errors = char_distance # Number of insertions, deletions, or substitutions
    total_chars = len(ground_truth)

    if total_chars == 0:
        return 100.0 # Handle empty ground truth

    cer = (char_errors / total_chars) * 100
    return 100.0 - cer # Return accuracy as a percentage

# Example usage:
ground_truth_text = "xxx"
ocr_output_text = "XXx" # Example error

accuracy = calculate_ocr_accuracy(ocr_output_text, ground_truth_text)
print(f"OCR Accuracy: {accuracy:.2f}%")

# The code above calculates accuracy directly, it can be wrapped it in a loop over dataset for a comprehensive evaluation
```

```python
# Testing processing a hand_written note, that is in image form

import base64

def encode_image(image_path):
    """Encoding the image to base64."""
    try:
        with open(image_path, "rb") as image_file:
            return base64.b64encode(image_file.read()).decode('utf-8')
    except FileNotFoundError:
        print(f"Error: The file {image_path} was not found.")
        return None
    except Exception as e:  # Added general exception handling
        print(f"Error: {e}")
        return None

image_path = "hand_written.png"

base64_image = encode_image(image_path)

ocr_response = client.ocr.process(
    model="mistral-ocr-latest",
    document={
        "type": "image_url",
        "image_url": f"data:image/jpeg;base64,{base64_image}" 
    },
    include_image_base64=True
)
```
